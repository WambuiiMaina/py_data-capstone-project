```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


```

#### Prevalence by region
What is the prevalence of cervical cancer and where are the regional hotspots


```python
prevalence_by_region = pd.read_csv('Prevalence_by_region.csv')
prevalence_by_region
```

```python
prevalence_by_region.info()
```

```python
screening_rate = prevalence_by_region['Prevalence'].mean()
screening_rate
```

```python
plt.barh(prevalence_by_region['Region'], prevalence_by_region['Prevalence'])

plt.title('Prevalence of Cervical Cancer by Region')
plt.xlabel('Prevalence (%)')
plt.ylabel('Region')

plt.tight_layout()
plt.show()
```

#### Prevalence by sociodemographic factors eg age, marital status,education, SES etc

Defining who is being screened and who is not and exposing the gaps that can be addressed

```python
sociodemographic_factors = pd.read_csv('Sociodemographic_data.csv')
sociodemographic_factors
```

```python
sociodemographic_factors.info()
```

#### Screening trend by age
Insights- it increases with age 

```python
age = sociodemographic_factors[
    sociodemographic_factors['Category'] == 'Age'
]

plt.plot(age['Group'], age['Yes_Percent'], marker='v', linestyle='-', color='purple')

plt.title("Trend of Screening Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Screening Rate (%)")

plt.grid(False)
plt.show()
```

#### Marital status

```python
marital_data = sociodemographic_factors[
    sociodemographic_factors['Category'] == 'Marital Status'
]

Fig, ax = plt.subplots()
ax.bar(marital_data['Group'], marital_data['Yes_Percent'], color=['skyblue', 'rosybrown', 'pink'])

plt.title("Screening Rate by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Screening Rate (%)")

plt.tight_layout()

plt.show()
```

#### Education level

```python
edu_data = sociodemographic_factors[
    sociodemographic_factors['Category'] == 'Education'
]

Fig, ax = plt.subplots()
ax.bar(edu_data['Group'], edu_data['Yes_Percent'], 
       color=['cyan',
              'teal',
              'pink',
              'purple'])

plt.title("Screening Rate by Education Level")

plt.xlabel("Education Level")
plt.ylabel("Screening Rate (%)")

plt.show()
```

#### Empolyment status

```python
employment_data = sociodemographic_factors[
    sociodemographic_factors['Category'] == 'Employment Status'
]

plt.pie(employment_data['Yes_Percent'], labels=employment_data['Group'], autopct='%1.1f%%', startangle=140)

plt.title("Screening Rate by Employment Status")

plt.show()
```

#### Wealth index

```python
wealth_data = sociodemographic_factors[
    sociodemographic_factors['Category'] == 'Wealth Index'
]
Fig, ax = plt.subplots()
ax.bar(wealth_data['Group'], wealth_data['Yes_Percent'], color=['lightcoral', 'lightseagreen', 'lightsteelblue', ])

plt.title("Screening Rate by Wealth Index")

plt.xlabel("Wealth Category")
plt.ylabel("Screening Rate (%)")
plt.show()
```

#### Residence pattern

```python
residence_data = sociodemographic_factors[
    sociodemographic_factors['Category'] == 'Residence'
]

plt.pie(residence_data['Yes_Percent'], labels=residence_data['Group'], autopct='%1.1f%%', startangle=140)

plt.title("Screening Rate by Residence")
plt.legend(title='Residence Type', labels=['Urban', 'Rural'])

plt.show()
```

#### Media exposure

```python
media_data = sociodemographic_factors[
    sociodemographic_factors['Category'] == 'Media Exposure'
]

plt.pie(media_data['Yes_Percent'], labels=media_data['Group'], autopct='%1.1f%%', startangle=140)

plt.title("Screening Rate by Media Exposure")

plt.show()
```

#### Summary of all the sociodemographic factors

```python
pivot = sociodemographic_factors.pivot(
    index='Category',
    columns='Group',
    values='Yes_Percent'
)

plt.figure(figsize=(12, 6))  # 👈 makes it bigger

sns.heatmap(
    pivot,
    annot=True,
    cmap='coolwarm',
    fmt=".1f",
    linewidths=0.5,
    cbar_kws={'label': 'Screening Rate (%)'}
)

plt.title("Cervical Cancer Screening Rates by Category and Group", fontsize=14)
plt.xlabel("Group")
plt.ylabel("Category")

plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```
